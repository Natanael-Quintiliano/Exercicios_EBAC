# Tarefa 1

### 1. Idade media das mães e dos pais por município (coluna munResNome)


```python
pd.set_option('display.precision', 2)
df_idade_media = df_sinasc[['IDADEPAI', 'IDADEMAE','munResNome']].groupby('munResNome').mean()
df_idade_media
```

### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
df_peso_medio = df_sinasc[['ESCMAE', 'DTNASC','SEXO','PESO']]
df_peso_medio = df_peso_medio[df_peso_medio['DTNASC'] == '2019-07-03'].groupby(['SEXO', 'ESCMAE'])
df_peso_medio.agg({'PESO'  :[('Min', 'min'),('Média', 'mean'), ('Max', 'max')]})
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
"""
Primeiro verificamos quais municipio teve menos nascimentos
"""
df_sinasc['munResNome'].value_counts()
```

```python
"""
Verificamos que o municipio de castanheira teve menos nascimentos filtramos a base agora
"""
df_menos_nasc = df_sinasc[[ 'IDADEMAE','IDADEPAI','munResNome']]
df_menos_nasc = df_menos_nasc[df_menos_nasc['munResNome'] == 'Castanheiras'].groupby('munResNome')
```

```python
"""
Apos isso vericamos: 
- qual a idade media, maxima, minima das maes nesse municipio?
- qual a idade media, maxima, minima dos pais nesse municipio?
"""

pd.set_option('display.precision', 2)
df_menos_nasc.agg({
                 'IDADEPAI'  :[('Min', 'min'), ('Média', 'mean'), ('Max', 'max')]
                 ,'IDADEMAE'  :[('Min', 'min'), ('Média', 'mean'), ('Max', 'max')]
                 })
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
"""
Primeiro filtramos as colunas necessarias
"""
df_nasc_marco = df_sinasc[['DTNASC','QTDFILVIVO','QTDFILMORT','IDADEPAI','IDADEMAE','munResNome']]
```

```python
"""
Depois ajustamos e filtramos os nascidos em março e ja verificamos qual municipio teve mais nascimentos
"""
df_nasc_marco['DTNASC'] = pd.to_datetime(df_nasc_marco['DTNASC'])
df_nasc_marco = df_nasc_marco[df_nasc_marco['DTNASC'].dt.month == 3]
df_nasc_marco['munResNome'].value_counts().idxmax()
```

```python
"""
Apos isso filtramos apenas os nascidos naquele municipio e vericamos: 
- qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
- qual a idade media, maxima, minima dos pais nesse municipio??
"""
pd.set_option('display.precision', 2)
df_nasc_marco = df_nasc_marco[df_nasc_marco['munResNome'] == 'Porto Velho'].groupby('munResNome')
df_nasc_marco.agg({
                  'QTDFILVIVO':[('Min', 'min'), ('Média', 'mean'), ('Max', 'max')]
                 ,'QTDFILMORT':[('Min', 'min'), ('Média', 'mean'), ('Max', 'max')]
                 ,'IDADEPAI'  :[('Min', 'min'), ('Média', 'mean'), ('Max', 'max')]
                 ,'IDADEMAE'  :[('Min', 'min'), ('Média', 'mean'), ('Max', 'max')]
                 })
```
