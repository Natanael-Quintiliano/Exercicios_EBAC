```python
"""
Importando as bibliotecas
"""
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Criando Graficos


### 1. Carregando a base <a name="lendo"></a>

<div style="text-align: right"
     
[Voltar ao índice](#questoes)

A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.     

```python
"""
Carregando Base
"""
df_sinasc = pd.read_csv('SINASC_RO_2019.csv')
```

```python
"""
Verificando missing values
"""
df_sinasc_sel = df_sinasc[['APGAR1','APGAR5','SEXO','GESTACAO','ESTCIVMAE','CONSULTAS']]
df_sinasc_sel.isna().sum()
```

```python
"""
Tratando missing values
"""
df_sinasc_sel.dropna(subset =['APGAR1'], inplace = True)
df_sinasc_sel.isna().sum()

df_sinasc_sel.dropna(subset =['APGAR5'], inplace = True)
df_sinasc_sel.isna().sum()

df_sinasc_sel.dropna(subset =['SEXO'], inplace = True)
df_sinasc_sel.isna().sum()

df_sinasc_sel['GESTACAO'].fillna(9, inplace = True) 
df_sinasc_sel.isna().sum()

df_sinasc_sel['ESTCIVMAE'].fillna(9, inplace = True) 
df_sinasc_sel.isna().sum()

df_sinasc_sel.isna().sum()
```

###  2. Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE. <a name="graficos"></a>

<div style="text-align: right"
     
[Voltar ao índice](#questoes)

```python
plt.close('all')
plt.rc('figure', figsize=(15, 10))
fig, axes = plt.subplots(2, 3)

a = sns.countplot(ax = axes[0, 0], x='APGAR1', data=df_sinasc_sel)
b = sns.countplot(ax = axes[0, 1], x='APGAR5', data=df_sinasc_sel)
c = sns.countplot(ax = axes[0, 2], x='ESTCIVMAE', data=df_sinasc_sel)
d = sns.countplot(ax = axes[1, 0], x='SEXO', data=df_sinasc_sel)
e = sns.countplot(ax = axes[1, 1], x='GESTACAO', data=df_sinasc_sel)

a.set_xticklabels(a.get_xticklabels(), rotation=90)
b.set_xticklabels(b.get_xticklabels(), rotation=90)
c.set_xticklabels(c.get_xticklabels(), rotation=90)
d.set_xticklabels(d.get_xticklabels(), rotation=90)
e.set_xticklabels(e.get_xticklabels(), rotation=90)

plt.subplots_adjust(wspace=1, hspace=1)
```
