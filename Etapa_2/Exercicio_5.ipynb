2. Quais são os três *status* das solicitações mais frequentes na base? Quais são suas frequências?
3. Construa uma nova variável que contenha o ano do **status**. Observe que ```data_status``` vem como tipo *object* no **DataFrame**. Dica: você pode usar o método ```.str``` para transformar o tipo da variável em string, em seguida um método como [**slice()**](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.slice.html) ou [**split()**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.str.split.html).

````python
"""
1) Repita os mesmos passos feitos em aula, mas selecione os PVLs realizados por municípios no estado do Rio de Janeiro.

Escolhi no site e ele ja me retornou os dados do Estado do Rio de Janeiro. Ja rodei a API e construi um Dataframe.
"""
url = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=RJ&tipo_interessado=Munic%C3%ADpio'

requisicao = requests.get(url)

dados_json = requisicao.json()

df_pvl_rj = pd.DataFrame(dados_json['items'])

df_pvl_rj.head(11)
````

````python
"""
2) Quais são os três status das solicitações mais frequentes na base? Quais são suas frequências?

Filtrei o Dataframe por 'status' e contei, mostrei todos por ter poucas variações.
"""
df_pvl_rj['status'].value_counts().head()
````

````python
"""
3) Construa uma nova variável que contenha o ano do status. Observe que data_status vem como tipo object no DataFrame. 
Dica: você pode usar o método .str para transformar o tipo da variável em string, em seguida um método como slice() ou split().

Criei uma nova coluna chamada 'ano' e utilizei o metodo slice paa retirar apenas o ano da coluna 'data_status'
"""
df_pvl_rj['ano'] = df_pvl_rj['data_status'].str.slice(start=6, stop=10)

df_pvl_rj.head(10)
````

# 2. Melhorando a interação com o API <a name="melhorando"></a>

<div style="text-align: right"
     
[Voltar ao índice](#questoes)

Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
"""
1) Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado ('Estado'ou Município),
e que devolve os dados da consulta no formato DataFrame.

Recriei tudo novamente apenas deixei a url sem filtros, após isso criei uma função que pega o estado e o tipo interessado e 
retorna um Dataframe
"""
url = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl' 

requisicao = requests.get(url)

dados_json = requisicao.json()

df_pvl = pd.DataFrame(dados_json['items'])

def consulta_dados(uf, tipo_interessado):
    dados_filtrados = df_pvl[(df_pvl['uf'] == uf) 
                      & (df_pvl['tipo_interessado'] == tipo_interessado)]
    return dados_filtrados
````

````python
"""
2) Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com status em 'Arquivado por decurso de prazo'
estão registradas?

Utilizando a função criada acima apenas filtrei e contei os resultados conforme solicitado.
"""
uf = 'MG'

tipo_interessado = 'Estado'

resultado_consulta = consulta_dados(uf, tipo_interessado)

resultado_consulta['status'].value_counts()
````

````python
"""
3) Qual é o município da Bahia com mais solicitações deferidas?

Utilizando a função criada acima apenas filtrei e contei os resultados conforme solicitado.
"""
uf = 'BA'

tipo_interessado = 'Município'

resultado_consulta = consulta_dados(uf, tipo_interessado)

df_deferidos = resultado_consulta[resultado_consulta['status'] == 'Deferido']

df_deferidos['interessado'].value_counts().head(10)
````

````python
"""
4) Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'
"""
uf = 'BA'

tipo_interessado = 'Estado'

resultado_consulta = consulta_dados(uf, tipo_interessado)

resultado_consulta.to_csv('Consulta Bahia.csv')
````
