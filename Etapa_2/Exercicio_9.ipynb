```python
"""
Importando as bibliotecas
"""
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Boxplot e Histograma

### 1. Na base do SINASC do DataSUS, faça um boxplot da idade da mãe. <a name="lendo"></a>

<div style="text-align: right"
     
[Voltar ao índice](#questoes)   

```python
"""
Carregando Base
"""
df_sinasc = pd.read_csv('SINASC_RO_2019.csv')
```

```python
"""
Verificando missing values
"""
df_sinasc_sel = df_sinasc[['IDADEMAE','PESO']]
df_sinasc_sel.isna().sum()
```

```python
sns.boxplot(y='IDADEMAE', data=df_sinasc_sel)
```

###  2. Faça um boxplot do peso do bebê. <a name="graficos"></a>

<div style="text-align: right"
     
[Voltar ao índice](#questoes)

```python
sns.boxplot(y='PESO', data=df_sinasc_sel)
```

###  3. Faça um histograma da idade da mãe. <a name="graficosII"></a>

<div style="text-align: right"
     
[Voltar ao índice](#questoes)  

```python
plt.close('all')
plt.rc('figure', figsize=(15, 10))
a = sns.countplot( x='IDADEMAE', data=df_sinasc_sel)
a.set_xticklabels(a.get_xticklabels(), rotation=90)
```

###  4. Faça um histograma do peso do bebê. <a name="graficosIII"></a>

<div style="text-align: right"
     
[Voltar ao índice](#questoes)

```python
plt.close('all')
plt.rc('figure', figsize=(15, 10))
a = sns.countplot( x='PESO', data=df_sinasc_sel)
a.set_xticklabels(a.get_xticklabels(), rotation=90)
```
