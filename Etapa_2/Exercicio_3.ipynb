# 1. Crie uma série do pandas a partir de uma lista com os dados abaixo: <a name="lista"></a>

<div style="text-align: right"
     
[Voltar ao índice](#questoes)

Em um estudo sobre alteração na tempreatura global, A NASA disponibiliza dados de diferenças de de temperatura média da superfície terrestre relativos às médias de temperatura entre 1951 e 1980. Os dados originais podem ser vistos no site da NASA/GISS, e estão dispostos a cada década na tabela abaixo.

|ano|anomalia termica|
|:-:|:----:|
| 1900 | -0.08 |
| 1920 | -0.27 |
| 1940 | 0.12 |
| 1960 | -0.03 |
| 1980 | 0.26 |
| 2000 | 0.40 |
| 2020 | 1.02 |

Crie uma séries do Pandas a partir de uma lista com esses dados.

```python
"""
Criando as lista ano e temperatura
"""
temperatura_lista = [-0.08,-0.27,0.12,-0.03,0.26,0.40,1.02]
ano_lista = [1900,1920,1940,1960,1980,2000,2020]

"""
Criando uma series a partir de uma lista
"""
temperatura_series = pd.Series(temperatura_lista)
temperatura_series
```

# 2. Coloque os anos nos índices conforme a tabela. <a name="indice"></a>

<div style="text-align: right"
     
[Voltar ao índice](#questoes)

```python
"""
Colocando os anos como indices
"""
ano_temperatura_series = pd.Series(temperatura_lista,
                                   index=ano_lista)
ano_temperatura_series
```

# 3. A partir do dicionário abaixo, crie uma séries do Pandas: <a name="dicionario"></a>

<div style="text-align: right"

# 4. Transforme o ndarray abaixo em um dataframe.  <a name="ndarray"></a>

<div style="text-align: right"
     
[Voltar ao índice](#questoes)

O numpy é capaz de gerar arrays n-dimensionais com números pseudo-aleatórios de acordo com uma variedade de distribuições, como no exemplo abaixo. Transforme esse nd-array em um DataFrame.

```python
"""
Criando um ndarray e tranformando ele em um Dataframe
"""
arr = np.random.normal(100, 10, (20,3))
pd.DataFrame(arr)

"""
Criando um Dataframe usando ndarray
"""
df = pd.DataFrame(np.random.normal(100, 10, (20,3)))
```

# 5. Nomeie os índices das linhas com inteiros de 1 a 20, e as colunas com os nomes "x1", "x2", e "x3" respectivamente. <a name="linhas"></a>

<div style="text-align: right"
     
[Voltar ao índice](#questoes)

```python
# Renomeando as colunas 

df.columns  = ['x1','x2','x3']
print(df)

# Renomeando os indices

df = df.rename(index=lambda x: x + 1)
print(df)
```

# 6. No DataFrame do exercício 5, crie uma nova coluna como sendo a média das três colunas, e dê a ela o nome de "media" (não recomendo colocar acentos em nomes de variáveis). <a name="media"></a>

<div style="text-align: right"
     
[Voltar ao índice](#questoes)

```python
"""
Adicionando nova coluna 'media', que é a soma das colunas 'x1','x2'e 'x3' dividas por 3 (quantidade de colunas)
"""
df['media'] = (df['x1'] + df['x2'] + df['x3']) / 3
df
```

# 7) No DataFrame do exercício 6, crie uma nova coluna chamada "log_med", contendo o logaritmo natural da média calculada no exercício 6  <a name="log"></a>

<div style="text-align: right"
     
[Voltar ao índice](#questoes)

```python
"""
Criando uma nova coluna 'log_med' contendo o logaritmo natural da média calculada no exercício 6
"""
df['log_med'] = np.log(df['media'])
df   
```
